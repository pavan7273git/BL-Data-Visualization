5. Write a Python program to plot two or more lines on same plot with suitable legends of each
line.

```python
import matplotlib.pyplot as plt


x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]  
y2 = [25, 20, 15, 10, 5]  


plt.plot(x, y1, label='y = x^2', color='b', marker='o')  
plt.plot(x, y2, label='y = 30 - x^2', color='r', marker='s')  


plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Multiple Line Plot with Legends')


plt.legend()


 #plt.grid(True)
plt.show()
```

6. Write a Python program to plot two or more lines with legends, different widths and colors.

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y1=[5,10,15,20,25]
y2=[4,8,12,16,20]

plt.plot(x,y1,label='y1=x*5',color='b',linestyle='dotted',marker='^',linewidth=2.0)
plt.plot(x,y2,label='y2=x*4',color='g',linestyle='dotted',marker='o',linewidth=3.0)

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Different ys')

plt.legend(loc='lower right')

plt.show()
```

7. Write a Python program to plot two or more lines with different styles


```python
import matplotlib.pyplot as plt


x=[1,2,3,4,5]
y1=[3,6,9,12,15]
y2=[15,12,9,6,3]
y3=[10,9,9.5,7,9]

plt.plot(x,y1,label='y1=x*3',color='r',linestyle='dotted',marker='s',linewidth=2.5)
plt.plot(x,y2,label='y2=reverse(y1)',color='b',linestyle='dashdot',marker='d',linewidth=3.0)
plt.plot(x,y3,label='y3=custom',color='g',linestyle='solid',marker='+',linewidth=3.5)

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Different styles lines')

plt.legend()

plt.show()
```

11. Write a Python program to plot several lines with different format styles in one command
using arrays.

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,2,3,4,5])
y1=np.array([3,6,9,12,15])
y2=np.array([15,12,9,6,3])
y3=np.array([10,9,9.5,7,9])

plt.plot(x,y1,label='y1=x*3',color='r',linestyle='dotted',marker='s',linewidth=2.5)
plt.plot(x,y2,label='y2=reverse(y1)',color='b',linestyle='dashdot',marker='d',linewidth=3.0)
plt.plot(x,y3,label='y3=custom',color='g',linestyle='solid',marker='+',linewidth=3.5)

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Different styles lines')

plt.legend()

plt.show()
```

12. Write a Python program to create multiple types of charts

```python
import matplotlib.pyplot as plt
import numpy as np
#import seaborn as sns


x = np.arange(1, 6)
y = [10, 20, 15, 25, 30]
z = [5, 15, 10, 20, 25]


fig, axs = plt.subplots(2, 2, figsize=(10, 8))


axs[0, 0].plot(x, y, marker='o', color='b', linestyle='-')
axs[0, 0].set_title("Line Chart")


axs[0, 1].bar(x, y, color='g')
axs[0, 1].set_title("Bar Chart")


axs[1, 0].scatter(x, y, color='r')
axs[1, 0].set_title("Scatter Plot")


axs[1, 1].hist(z, bins=5, color='purple', edgecolor='black')
axs[1, 1].set_title("Histogram")


plt.tight_layout()


plt.show()
```
