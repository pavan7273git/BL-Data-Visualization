1. Write a Python programming to display a bar chart of the popularity of programming

7. Write a Python programming to display a bar chart of the popularity of programming
Languages. Specify the position of each bar plot.
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
import matplotlib.pyplot as plt
import numpy as np


languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]


x_pos = np.arange(len(languages))


plt.bar(x_pos, popularity, color=['blue', 'green', 'red', 'purple', 'orange', 'cyan'])


plt.xlabel("Programming Languages")
plt.ylabel("Popularity (%)")
plt.title("Popularity of Programming Languages")


plt.xticks(x_pos, languages)

plt.show()
```

9. Write a Python programming to display a bar chart of the popularity of programming

10. Write a Python program to create bar plot of scores by group and gender. Use multiple X
values on the same chart for men and women.
Sample Data:
Means (men) = (22, 30, 35, 35, 26)
Means (women) = (25, 32, 30, 35, 29)

```python
import matplotlib.pyplot as plt
import numpy as np


groups = ['G1', 'G2', 'G3', 'G4', 'G5']
means_men = [22, 30, 35, 35, 26]
means_women = [25, 32, 30, 35, 29]


x = np.arange(len(groups))  # [0, 1, 2, 3, 4]
width = 0.3 


plt.bar(x - width/2, means_men, width, label='Men', color='blue', alpha=0.7)
plt.bar(x + width/2, means_women, width, label='Women', color='pink', alpha=0.7)


plt.xlabel("Groups")
plt.ylabel("Scores")
plt.title("Scores by Group and Gender")
plt.xticks(x, groups) 
plt.legend()

plt.show()
```

11. Write a Python program to create bar plot from a DataFrame.
Sample Data Frame:
a b c d e
2 4,8,5,7,6
4 2,3,4,2,6
6 4,7,4,7,8
8 2,6,4,8,6
10 2,4,3,3,2

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create the DataFrame
data = {
    'a': [2, 4, 6, 8, 10],
    'b': [4, 2, 4, 2, 6],
    'c': [8, 3, 7, 6, 4],
    'd': [5, 4, 4, 4, 3],
    'e': [7, 2, 7, 8, 3],
}

df = pd.DataFrame(data)

# Set 'a' column as index (optional)
df.set_index('a', inplace=True)

# Plot bar chart
df.plot(kind='bar', figsize=(8, 6))

# Labels and title
plt.xlabel("Index")
plt.ylabel("Values")
plt.title("Bar Plot from DataFrame")

# Show plot
plt.show()
```

12. Write a Python program to create bar plots with error bars on the same figure. Sample Date
Mean velocity: 0.2474, 0.1235, 0.1737, 0.1824
Standard deviation of velocity: 0.3314, 0.2278, 0.2836, 0.2645

```python
import matplotlib.pyplot as plt
import numpy as np


mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824] 
std_dev = [0.3314, 0.2278, 0.2836, 0.2645] 


categories = ['A', 'B', 'C', 'D']


x_pos = np.arange(len(categories))


plt.bar(x_pos, mean_velocity, yerr=std_dev, capsize=5, color='blue', alpha=0.7, label='Mean Velocity')


plt.xticks(x_pos, categories)
plt.xlabel("Categories")
plt.ylabel("Mean Velocity")
plt.title("Bar Plot with Error Bars")
plt.legend()


plt.show()
```

13. Write a Python program to create bar plots with errorbars on the same figure. Attach a text
label above each bar displaying men means (integer value).
Sample Data
Mean velocity: 0.2474, 0.1235, 0.1737, 0.1824
Standard deviation of velocity: 0.3314, 0.2278, 0.2836, 0.2645


```python
import matplotlib.pyplot as plt
import numpy as np


mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824] 
std_dev = [0.3314, 0.2278, 0.2836, 0.2645]  
categories = ['A', 'B', 'C', 'D'] 


x_pos = np.arange(len(categories))

bars = plt.bar(x_pos, mean_velocity, yerr=std_dev, capsize=5, color='blue', alpha=0.7, label='Mean Velocity')


for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.02, f"{int(height)}", ha='center', va='bottom', fontsize=12, color='black')


plt.xticks(x_pos, categories)
plt.xlabel("Categories")
plt.ylabel("Mean Velocity")
plt.title("Bar Plot with Error Bars and Labels")
plt.legend()

plt.show()
```
