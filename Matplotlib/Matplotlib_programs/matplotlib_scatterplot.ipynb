```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

1. Write a Python program to draw a scatter graph taking a random distribution in X and Y and
plotted against each other.

```python
x=np.random.rand(50)
y=np.random.rand(50)

plt.scatter(x,y,color='b',alpha=0.5,edgecolor='r',label='Group 1 data')
plt.legend()
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Graph with random values')
plt.show()
```

2. Write a Python program to draw a scatter plot with empty circles taking a random distribution
in X and Y and plotted against each other.

```python
x=np.random.rand(50)
y=np.random.rand(50)

plt.scatter(x,y,color='b',edgecolor='r',facecolors='none',s=45)
plt.legend()
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Graph with random values')
plt.show()
```

3. Write a Python program to draw a scatter plot using random distributions to generate balls of
different sizes

```python
x=np.random.rand(50)
y=np.random.rand(50)

sizes=np.random.rand(50)*100
plt.scatter(x,y,color='blue',edgecolors='black',s=sizes)
plt.legend()
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Graph with random values')
plt.show()
```

5. Write a Python program to draw a scatter plot for three different groups camparing weights
and heights.

```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(42)
weights_group1 = np.random.randint(50, 80, 10)
heights_group1 = np.random.randint(150, 180, 10)

weights_group2 = np.random.randint(60, 90, 10)
heights_group2 = np.random.randint(160, 190, 10)

weights_group3 = np.random.randint(70, 100, 10)
heights_group3 = np.random.randint(170, 200, 10)


plt.scatter(weights_group1, heights_group1, color='red', label='Group 1', alpha=0.6, edgecolors='black')
plt.scatter(weights_group2, heights_group2, color='blue', label='Group 2', alpha=0.6, edgecolors='black')
plt.scatter(weights_group3, heights_group3, color='green', label='Group 3', alpha=0.6, edgecolors='black')

plt.xlabel('Weight (kg)')
plt.ylabel('Height (cm)')
plt.title('Scatter Plot of Weights vs Heights for Three Groups')
plt.legend() 
plt.show()
```
